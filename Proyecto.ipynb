# **Vamos a sumar todas las columnas de tráfico de datos para calcular cuanto se consumió de datos en pandemia en el año y en los meses**

```python
##Agrupamos las columnas en un dataframe

trafico_columnas=['trafico datos: internacional (gb)',
  'trafico datos: naps - colombia (gb)',
  'trafico datos: acuerdos de transito o peering directo (gb)',
  'trafico datos: local (gb)'  ]

##Agruparemos por años y sumaremos cada categoría
suma_anual=df.groupby('año',as_index=False)[trafico_columnas].sum()
suma_anual
```

```python
## Acontinuación tendremos los datos agrupados con los años en los cuales se gasto los datos de internet
suma_anual['uso_total_gb']=suma_anual[trafico_columnas].sum(axis=1)
suma_anual[['año','uso_total_gb']]
```

```python
## y ahora visualizamos la gráfica
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6,4))
sns.barplot(
    data=suma_anual,
    x='año',
    y='uso_total_gb',
    color='blue'
)
plt.xlabel('Año')
plt.ylabel('Uso total de datos (GB)')
plt.title('Uso total de datos en el año')
plt.show()
```

# **Ahora miraremos el comportamiento en el mes**

```python
##creamos dos dataframe donde guardaremos datos del año 2021 y 2022
df_2020=df[df['año']==2020]
df_2021=df[df['año']==2021]
```

```python
#ahora vamos a agrupar en otros dos dataframe los meses

meses_2020=df_2020.groupby('mes', as_index=False)[trafico_columnas].sum()
meses_2021=df_2021.groupby('mes', as_index=False)[trafico_columnas].sum()
meses_2020
```

```python
meses_2020['Suma_mensual_total']=meses_2020[trafico_columnas].sum(axis=1)
meses_2020

mapping_meses = {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}
meses_2020['nombre_mes'] = meses_2020['mes'].map(mapping_meses)
meses_2020
```

```python
##Ahora vamos a graficar
plt.figure(figsize=(6,4))
sns.barplot(data=meses_2020, x='nombre_mes', y='Suma_mensual_total', color='blue')
plt.xlabel('Mes')
plt.ylabel('Uso total de datos (GB)')
plt.title('Uso total de datos en el año 2020')
plt.xticks(rotation=90)
plt.show()
```

```python
##Consumo mensual año 2021
meses_2021['Suma_mensual_total']=meses_2021[trafico_columnas].sum(axis=1)

meses_2021['nombre_mes'] = meses_2021['mes'].map(mapping_meses)
meses_2021
```

```python
#Haremos la  tabla de contingencia

from scipy.stats import chi2_contingency ## Tabla de contingencias
from scipy.stats import norm, kurtosis ##Curtosis

tabla_contingencia=pd.crosstab(df['año'],df['proveedor'])
tabla_contingencia
```

```python
print("Tabla de contingencia (frecuencias):")
print(tabla_contingencia)
```

```python
chi2, p_value, grados_libertad, frecuencias_esperadas = chi2_contingency(tabla_contingencia)
significancia= p_value<0.05
print(f"Chi2 = {chi2:.3f}")
print(f"significancia = {significancia}")
print(f"Grados de libertad = {grados_libertad}\n")
```
