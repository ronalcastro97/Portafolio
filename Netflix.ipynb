```python
dataset['Release_Date']=pd.to_datetime(dataset['Release_Date'].str.strip())
dataset
```

```python
dataset['año']=dataset['Release_Date'].dt.year
titulos_por_año=dataset.groupby('año')['Title'].count() #me crea un dataset que separa por filas los años con sus respectivas películas
titulos_por_año
```

```python

plt.figure(figsize=(12,6))
titulos_por_año.plot(kind='bar', color='skyblue')
plt.title('Cantidad de lanzamientos por año')
plt.xlabel('Año')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
max_año=titulos_por_año.idxmax()
numero_peliculas=titulos_por_año.max()
print(f"El año con el mayor número de títulos es {max_año} con {numero_peliculas} títulos.")
```

```python
##9) What are the different Ratings defined by Netflix ?
#Q. 9) What are the different Ratings defined by Netflix ?
#Q. 9.1) How many Movies got the 'TV-14' rating, in Canada ?
#Q. 9.2) How many TV Shows got the 'R' rating, after year 2018 ?

print(dataset['Rating'].unique()) ##lista con las clasificaciones

tv_14=dataset[(dataset['Rating']=='TV-14') & (dataset['Country']=='Canada')]
print(f"En canada hay {tv_14.shape[0]} de categoria TV.14")

r=dataset[(dataset['Rating']=='R') & (dataset['año']>2018)]
print(f"En canada hay {r.shape[0]} de categoria R")


```

```python
#filtramos las películas con duración en minutos


dataset['Duracion_min']=0
dataset
for i in range(len(dataset)):
  valor=dataset.loc[i, 'Duration']

  if isinstance(valor, str):
    if 'min' in valor:
      dataset.loc[i, 'Duracion_min']=int(valor.split()[0]) #Divide el texto por espacio y toma la primera parte.
minutos_pelicula= dataset['Duracion_min'].max()
titulo_pelicula = dataset['Title'][dataset['Duracion_min'] == minutos_pelicula].values[0]


print(f"La película con mayor duración es {minutos_pelicula} y se llama {titulo_pelicula}")


```

```python
##ahora filtraremos la serie
dataset['duracion_max_tvshow']=0


for i in range(len(dataset)):
  valor=dataset.loc[i, 'Duration']

  if isinstance(valor, str):
    if 'Season' in valor:
      dataset.loc[i, 'duracion_max_tvshow']=int(valor.split()[0])

##series con más temporadas
serie_temporadas= dataset['duracion_max_tvshow'].max()
titulo_serie = dataset['Title'][dataset['duracion_max_tvshow'] == serie_temporadas].values[0]

print(f"La serie con mayor duración es {serie_temporadas} temporadas y se llama {titulo_serie}")
```

```python
##11) Which individual country has the Highest No. of TV Shows ?
#miremos  el país que más tiene tv show
tvshows=dataset[dataset['Category']=='TV Show'].copy()
tvshows=tvshows.dropna(subset=['Country'])
tvshows['Country']=tvshows['Country'].str.split(',')
tvshows=tvshows.explode('Country')
conteo= tvshows.groupby('Country').size().sort_values(ascending=False)


print(f"El país con más de tv shows es {conteo.index[0]} con {conteo.iloc[0]} programas de televisión")
```

```python
plt.figure(figsize=(12, 6))
conteo.head(10).plot(kind='bar', color='rosybrown')
plt.title('Top 10 países con más TV Shows')
plt.xlabel('País')
plt.ylabel('Cantidad de TV Shows')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
